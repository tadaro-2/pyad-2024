#Данные

```python
import pandas as pd
import numpy as np
from surprise import Dataset, Reader, SVD, accuracy
from surprise.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_absolute_error
import pickle
from sklearn.model_selection import train_test_split
```

Работа со сдвигами:

```python
books["Year-Of-Publication"] = books["Year-Of-Publication"].astype(str)

column_names = books.columns.tolist()
start = column_names.index("Book-Author")

shifted = books[books["Year-Of-Publication"].str.match("[^0-9]", na=False)]

for i in shifted.index:
    row_data = books.loc[i]
    print(f"До изменения (строка {i}): {row_data[column_names]}")

    for j in range(len(column_names) - 1, start, -1):
        row_data[column_names[j]] = row_data[column_names[j - 1]]

    row_data[column_names[start]] = np.nan
    books.loc[i] = row_data

    print(f"После изменения (строка {i}): {row_data[column_names]}")
```

Пропущеные ячейки:

```python
books = books.dropna()
```

Несуществующие года:

```python
books["Year-Of-Publication"] = pd.to_numeric(books["Year-Of-Publication"], errors="coerce")
books["Year-Of-Publication"] = books["Year-Of-Publication"].astype(int)
books = books[books["Year-Of-Publication"] <= 2025]
```

```python
print(books["Year-Of-Publication"].value_counts().sort_values(ascending=True).head(10))
```

Картинки:

```python
books = books.drop(columns=['Image-URL-S', 'Image-URL-M', 'Image-URL-L'], errors='ignore')
```

```python
books.head()
```

Не учитываем 0 рейтинг:

```python
ratings = ratings[ratings['Book-Rating'] != 0]
```

Оценено один раз:

```python
books = books[books["ISBN"].isin(ratings["ISBN"])]
ratings = ratings.groupby("User-ID").filter(lambda x: len(x) > 1)
```

```python
ratings.head()
```

#Обучение SVD

```python
from surprise import Dataset, Reader, SVD, accuracy
from surprise.model_selection import train_test_split

reader = Reader(rating_scale=(1, 10))

data = Dataset.load_from_df(ratings[["User-ID", "ISBN", "Book-Rating"]], reader)

trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

svd = SVD()
svd.fit(trainset)

predictions = svd.test(testset)

mae = accuracy.mae(predictions)

with open("svd.pkl", "wb") as model_file:
    pickle.dump(svd, model_file)
```

# Обучение LinReg

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

avg_ratings = ratings.groupby("ISBN")["Book-Rating"].mean().reset_index()
avg_ratings.rename(columns={"Book-Rating": "Avg-Rating"}, inplace=True)

data = pd.merge(books, avg_ratings, on="ISBN", how="left").dropna(subset=["Avg-Rating"])

X = data[["Book-Title", "Book-Author", "Publisher", "Year-Of-Publication"]]
y = data["Avg-Rating"]

title_tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
cat_encoder = OneHotEncoder(handle_unknown='ignore')
num_scaler = StandardScaler()

preprocessor = ColumnTransformer(transformers=[
    ("title", title_tfidf, "Book-Title"),
    ("author", cat_encoder, ["Book-Author"]),
    ("publisher", cat_encoder, ["Publisher"]),
    ("year", num_scaler, ["Year-Of-Publication"])
])

model = Pipeline([
    ("preprocessor", preprocessor),
    ("regressor", SGDRegressor(random_state=42, max_iter=1000, tol=1e-3))
])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
print(mae)

with open("linreg.pkl", "wb") as file:
  pickle.dump(model, file)
```

# Рекомендации

```python
import pandas as pd
import joblib
import numpy as np

ratings_df = pd.read_csv("Ratings.csv")
books_df = pd.read_csv("Books.csv")

svd_model = joblib.load("svd.pkl")
linreg_model = joblib.load("linreg.pkl")

user_zero_ratings = ratings_df[ratings_df["Book-Rating"] == 0]
user_zero_counts = user_zero_ratings.groupby("User-ID").size()
user_with_most_zeros = user_zero_counts.idxmax()

user_zero_books = user_zero_ratings[user_zero_ratings["User-ID"] == user_with_most_zeros]["ISBN"].values

svd_predictions = svd_model.test([(user_with_most_zeros, book, 0) for book in user_zero_books])
high_rating_books = [pred.iid for pred in svd_predictions if pred.est >= 8]

books_for_linreg = books_df[books_df["ISBN"].isin(high_rating_books)].copy()
books_for_linreg["Predicted-Rating"] = linreg_model.predict(books_for_linreg[["Book-Title", "Book-Author", "Publisher", "Year-Of-Publication"]])

sorted_books = books_for_linreg.sort_values(by="Predicted-Rating", ascending=False)

top_recommendations = sorted_books[["ISBN", "Book-Title", "Book-Author", "Predicted-Rating"]].head(10)

print(top_recommendations)
```
